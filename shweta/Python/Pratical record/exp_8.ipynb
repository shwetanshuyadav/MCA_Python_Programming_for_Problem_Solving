```python
def categorize_files_by_extension(files):
    categorized = {}
    for file in files:
        # Extract file extension
        ext = file.split('.')[-1]
        # Add file to corresponding list in dictionary
        if ext not in categorized:
            categorized[ext] = []
        categorized[ext].append(file)
    return categorized

# Example input
files = ['report.pdf', 'data.csv', 'photo.jpg', 'summary.pdf', 'script.py']

# Categorize files
result = categorize_files_by_extension(files)
print(result)
```
