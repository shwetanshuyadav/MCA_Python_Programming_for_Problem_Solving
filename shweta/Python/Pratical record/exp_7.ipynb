```python
import string

def count_word_frequency(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = text.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    return frequency

text = "Rain rain go away, come again another day."
result = count_word_frequency(text)
print(result)
```
