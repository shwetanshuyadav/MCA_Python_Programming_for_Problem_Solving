```python
import pandas as pd
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer', 'Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer'],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories', 'Electronics', 'Electronics'],
    'Units_Sold': [30, 120, 85, 45, 25, 50, 60, 40, 55, 30],
    'Price_Per_Unit': [1200, 25, 45, 300, 150, 1150, 22, 48, 310, 145],
    'Month': ['Jan', 'Jan', 'Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Feb', 'Feb']
}
pd.DataFrame(data).to_csv('sales_dataset.csv', index=False)

df = pd.read_csv('sales_dataset.csv')
df['Total_Revenue'] = df['Units_Sold'] * df['Price_Per_Unit']

product_revenue = df.groupby('Product')['Total_Revenue'].sum().reset_index()
category_units = df.groupby('Category')['Units_Sold'].sum().reset_index()
category_avg_price = df.groupby('Category')['Price_Per_Unit'].mean().reset_index()
max_units_row = df.loc[df['Units_Sold'].idxmax()]
max_product = max_units_row['Product']
max_value = max_units_row['Units_Sold']
month_revenue = df.groupby('Month')['Total_Revenue'].sum().reset_index()
filtered_products = df[df['Units_Sold'] > 40]
sorted_df = df.sort_values(by='Total_Revenue', ascending=False)

# --- DISPLAYING RESULTS ---
print("Total Revenue per Product:\n", product_revenue)
print("\nTotal Units per Category:\n", category_units)
print("\nAverage Price per Category:\n", category_avg_price)
print(f"\nProduct with maximum units sold: {max_product} ({max_value} units)")
print("\nRevenue per Month:\n", month_revenue)
print("\nProducts with more than 40 units sold:\n", filtered_products)
print("\nSorted Dataset (by Revenue):\n", sorted_df)
```
