```python
def sort_employees_by_salary(employees):
    
    sorted_employees = sorted(employees, key=lambda x: x['salary'])
    return sorted_employees

employees = [
    {"name": "John", "salary": 45000},
    {"name": "Mary", "salary": 52000},
    {"name": "Alex", "salary": 48000}
]

sorted_list = sort_employees_by_salary(employees)
print("Sorted by Salary:", sorted_list)
```
