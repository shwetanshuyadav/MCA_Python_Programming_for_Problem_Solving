```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Employee_ID': range(1, 101),
    'Age': np.random.randint(22, 60, 100),
    'Experience': np.random.randint(1, 35, 100),
    'Salary': np.random.randint(30000, 150000, 100),
    'Department': np.random.choice(['IT', 'HR', 'Sales', 'Marketing', 'Finance'], 100)
}
data['Salary'] = data['Experience'] * 4000 + np.random.randint(20000, 50000, 100)

df = pd.DataFrame(data)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 12))

plt.subplot(3, 2, 1)
sns.histplot(df['Salary'], kde=True, color='skyblue')
plt.title('Distribution of Salary')

plt.subplot(3, 2, 2)
sns.boxplot(x=df['Age'], color='lightgreen')
plt.title('Box Plot of Employee Age')

plt.subplot(3, 2, 3)
sns.countplot(data=df, x='Department', palette='viridis')
plt.title('Employee Count by Department')

plt.subplot(3, 2, 4)
sns.scatterplot(data=df, x='Experience', y='Salary', hue='Department')
plt.title('Relationship: Experience vs Salary')

plt.subplot(3, 2, 5)
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
