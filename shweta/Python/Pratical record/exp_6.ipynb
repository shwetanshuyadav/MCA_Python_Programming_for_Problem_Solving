```python
def count_lines_words_chars(filename):
    lines = 0
    words = 0
    characters = 0

    with open(filename, 'r') as file:
        for line in file:
            lines += 1
            words += len(line.split())
            characters += len(line)

    print(f"Lines: {lines}")
    print(f"Words: {words}")
    print(f"Characters: {characters}")

# Replace 'sample.txt' with your text file name
filename = 'sample.txt'
count_lines_words_chars(filename)
```
