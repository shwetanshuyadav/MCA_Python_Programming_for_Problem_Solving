```python
def reverse_number_arithmetic(n):
    
    is_negative = n < 0
    num = abs(n)
    reversed_num = 0

    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num = num // 10
    return -reversed_num if is_negative else reversed_num
print(f"12345 reversed: {reverse_number_arithmetic(12345)}")
```
