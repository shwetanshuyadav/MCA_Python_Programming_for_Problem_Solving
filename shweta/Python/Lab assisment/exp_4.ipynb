```python
import statistics

data = [15, 18, 12, 15, 16, 18, 15, 17, 18, 16]
mean_value = statistics.mean(data)
median_value = statistics.median(data)
mode_values = statistics.multimode(data)

print(f"Data: {data}")
print("-" * 37)
print(f"Mean (Average): {mean_value}")
print(f"Median (Middle): {median_value}")
print(f"Mode(s) (Most Frequent): {mode_values}")
```
