```python
def reverse_vowels(input_string):
    vowels = "aeiouAEIOU"
    vowel_list = []
    for char in input_string:
        if char in vowels:
            vowel_list.append(char)

    vowel_list.reverse()
    result_list = []
    vowel_index = 0
   
    for char in input_string:
        if char in vowels:
            result_list.append(vowel_list[vowel_index])
            vowel_index += 1
        else:
            result_list.append(char)

    return "".join(result_list)

sentence = input("Enter sentence: ")
output_sentence = reverse_vowels(sentence)
print(f"Output: {output_sentence}")
```
