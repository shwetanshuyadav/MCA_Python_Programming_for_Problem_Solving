```python
# Probability of rolling a die (fair 6-sided die)
prob_die = 1 / 6

# Probability of flipping a coin (fair coin)
prob_coin = 1 / 2

print(f"Probability of getting any specific number on a die: {prob_die}")
print(f"Probability of getting heads or tails on a coin: {prob_coin}")

# Probability of both events happening together (independent events)
prob_both = prob_die * prob_coin
print(f"Probability of getting, for example, a 4 and heads: {prob_both}")
```
