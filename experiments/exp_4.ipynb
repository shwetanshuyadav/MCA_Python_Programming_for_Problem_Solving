```python
def search_item_in_list(lst, item):
    for index, value in enumerate(lst, start=1):
        if value == item:
            print(f"Item found at position {index}")
            return
    print("Item not found")
    
lst = [12, 45, 33, 21, 88]
item_to_search = 33

search_item_in_list(lst, item_to_search)
```
